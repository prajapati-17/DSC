```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns 
sns.set()
```

```python
data = pd.read_csv('real_estate_price_size.csv')
```

```python
data
```

```python
data.head(10)
```

```python
data.describe()
```

```python
y = data['size']
x1 = data['price']
```

```python
plt.scatter(x1,y)
plt.xlabel('price')
plt.ylabel('size')
plt.show()
```

```python
x=sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=0.0033*x1-122.3349
fig = plt.plot(x1,yhat,lw=2,c='red',label='regression line')
plt.xlabel('price')
plt.ylabel('size')
plt.show()
```
